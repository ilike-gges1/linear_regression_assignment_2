```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np #these are all necessary packages for the graph and the line
```

```python
data = pd.read_csv('regrex1.csv') #re-reading the csv, using pandas to make it more malleable for data analysis
```

```python
x_lr = np.array(data.x) #creating a variable out of an array of the x values in the CSV file
y_lr = np.array(data.y) #creating a variable out of an array of the y values in the CSV file

print(x_lr) 
print (y_lr) #This is just another way to print the CSV file
```

```python
plt.scatter(x_lr, y_lr, color='k') #creates the scatterplot of the data using matplotlib and the two created arrays for the data
plt.title('Scatter Plot Using Python') #title of the plot
plt.xlabel('the x value') #the title of the x axis
plt.ylabel('the y value') #the title of the y axis
plt.show()
```
